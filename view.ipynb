```python
file_path = './../DataSet/03_CHAOS/mri/t1_img/1_image_T1DUAL_OutPhase.nii.gz'
```

```python
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# Load data

data = nib.load(file_path).get_fdata()

def browse_images(slice_idx):
    plt.figure(figsize=(6, 6))
    # data.shape is (256, 256, 35). We slice the 3rd dimension.
    plt.imshow(data[:, :, slice_idx], cmap='gray')
    plt.axis('off')
    plt.title(f'Slice: {slice_idx}')
    plt.show()
```

```python
interact(browse_images, slice_idx=IntSlider(min=0, max=data.shape[2]-1, step=1, value=17))
```
