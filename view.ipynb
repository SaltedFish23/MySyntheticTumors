```python
file_path = 'output/imagesTr/3.nii.gz'
label_path = 'output/labelsTr/3.nii.gz'
```

```python
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider

# Load data

data = nib.load(file_path).get_fdata()
label = nib.load(label_path).get_fdata()

def browse_images(slice_idx):
    # Create a figure with 2 subplots (1 row, 2 columns)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot 1: The Raw Data
    ax1.imshow(data[:, :, slice_idx], cmap='gray')
    ax1.set_title(f'Image Slice: {slice_idx}')
    ax1.axis('off')
    
    # Plot 2: The Label
    # Using 'jet' or 'nipy_spectral' helps distinguish different label classes
    ax2.imshow(label[:, :, slice_idx], 
           cmap='tab10',  # 'tab20' if you have >10 classes
           vmin=0, 
           vmax=10,       # Set this to 10 (or 20) regardless of your max label
           interpolation='nearest')
    ax2.set_title(f'Label Slice: {slice_idx}')
    ax2.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
interact(browse_images, slice_idx=IntSlider(min=0, max=data.shape[2]-1, step=1, value=17))
```
